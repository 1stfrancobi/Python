```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Customer Call List.xlsx')
df
```

```python
df.duplicated()
```

```python
df = df.drop_duplicates()
df
```

```python
df.duplicated()
```

```python
df = df.drop(columns = 'Not_Useful_Column')
```

```python
df
```

```python
# Clean Last name

df['Last_Name'] = df['Last_Name'].str.strip('/._')
```

```python
df
```

```python
df.Phone_Number
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace(r'\D', '', regex = True)
```

```python
df
```

```python
df['Cleaned_Phone_Number'] = np.where(
    df['Phone_Number'].str.len() == 10,
    df['Phone_Number'].str.slice(0, 3) + '-' +
    df['Phone_Number'].str.slice(3, 6) + '-' +
    df['Phone_Number'].str.slice(6, 10),
    np.nan
)    
```

```python
df
```

```python
df = df.drop(columns = 'Phone_Number')
df
```

```python
# Display the phone number column without NaN

df['Cleaned_Phone_Number'] = df['Cleaned_Phone_Number'].fillna('')
```

```python
df
```

```python
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(pat = ',', n = 2, expand = True)
```

```python
df
```

```python
df = df.drop(columns = 'Address')
```

```python
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
```

```python
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
```

```python
df
```

```python
df = df.fillna('')
df
```

```python
for i in df.index:
    if df.loc[i, 'Do_Not_Contact'] == 'Y':
        df.drop(i, inplace = True)
```

```python
df
```

```python
for x in df.index:
    if df.loc[x, 'Cleaned_Phone_Number'] == '':
        df.drop(x, inplace = True)
```

```python
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.reset_index(drop = True)
```
